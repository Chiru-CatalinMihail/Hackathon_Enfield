```python
# Libraries needed for reading DICOM files
import pydicom
```

```python

file_path = './3000566.000000-NA-03192/1-039.dcm'


def read_dicom_file(file_path):
    '''
    Read a DICOM file and return the pixel data
    '''

    # Read the DICOM file
    dicom_file = pydicom.dcmread(file_path)

    # Get the pixel data
    pixel_data = dicom_file.pixel_array

    return pixel_data


```
